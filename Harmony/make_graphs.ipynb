```python
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm
import numpy as np

# c:\Windows\Fonts\TIMES.TTF c:\Windows\Fonts\TIMESBD.TTF c:\Windows\Fonts\TIMESBI.TTF c:\Windows\Fonts\TIMESI.TTF

kaust_colors_updated = ["#5284C4", "#00A6AA", "#F0B500", "#F18F00", "#CDCE00", "#9C6FAE"]
models = ["MAE", "iBOT", "CLIP", "SLIP", "MaskCLIP", "Harmony"]
data = {
    "MAE": [14.6, 26.9, 28.9],
    "iBOT": [58.2, 37.0, 44.9],
    "CLIP": [57.0, 35.3, 41.0],  
    "SLIP": [51.5, 35.0, 40.0],
    "MaskCLIP": [56.1, 34.5, 39.4],
    "Harmony": [64.4, 37.5, 45.6]
}

# Redefine the figure to include task names on the y-axis
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.15
index = np.arange(3)

# Plot each model's data with specific colors and task names on the y-axis only
for i, (model, scores) in enumerate(data.items()):
    adjusted_scores = [score if score is not None else 0 for score in scores]  # Replace None with 0 for plotting
    ax.barh(index + i * bar_width, adjusted_scores, bar_width, label=model, color=kaust_colors_updated[i % len(kaust_colors_updated)])

# Customize the plot
ax.set_xlabel('Performance')
ax.set_yticks(index + (bar_width * (len(models) - 1)) / 2)
ax.set_yticklabels(["Image Classification", "Semantic Segmentation", "Object Detection"])
ax.set_title('Performance of Models Across Tasks')
ax.legend(title="Models", bbox_to_anchor=(1.05, 1), loc='upper left')


plt.tight_layout()
plt.show()
```
